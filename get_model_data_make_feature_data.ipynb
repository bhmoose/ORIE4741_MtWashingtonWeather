```python
#Reset index of filtered model data
df_shared = df
df_shared.reset_index(inplace=True, drop=True)
```

```python
#Convert datetime to date (time is unnecessary)
df_shared['datetime'] = df_shared['datetime'].dt.date
```

```python
df_shared
```

```python
#Concatenate model features with station features
all_features = pd.concat([df_shared, mwobs_features], axis = 1)
all_features.drop(columns = ['date'], inplace=True)
```

```python
#Peform one-hot encoding
all_features = pd.concat([all_features.iloc[:, 0], pd.get_dummies(all_features.iloc[:, 1:])], axis = 1)
```

```python
all_features.columns.values
```

```python
#Add label to date column
newcols = ['date']
newcols.extend([i for i in all_features.columns[1:]])
all_features.columns = newcols

```

```python
#Convert date to year month day features
all_features.insert(0, 'year', pd.to_datetime(all_features['date']).dt.year)
all_features.insert(0, 'month', pd.to_datetime(all_features['date']).dt.month)
all_features.insert(0, 'day', pd.to_datetime(all_features['date']).dt.day)
all_features.drop(columns = ['date'], inplace=True)
```

```python
#Convert booleans to floats (0 or 1)
all_features = all_features.astype(float)
```

```python
all_features.columns.values
```

```python
#Find 11pm-5am change in all features, replace 11pm data with 6-hour 11pm-5am change
featurevars = ['M', 'BL', 'FG', 'RA', 'SN', 't', 'dt', 'rh', 'ws', 'vis', 'gust']
for i in range(len(featurevars)):
  featurename = featurevars[i] + '_11pm'
  all_features[featurename] = all_features[featurevars[i] + '_5am'] - all_features[featurename]

dummies_5am = ['wd_5am_E', 'wd_5am_SE', 'wd_5am_S', 'wd_5am_SW', 'wd_5am_W', 'wd_5am_NW', 'wd_5am_N', 'wd_5am_NE']
dummies_11pm = ['wd_11pm_E', 'wd_11pm_SE', 'wd_11pm_S', 'wd_11pm_SW', 'wd_11pm_W', 'wd_11pm_NW', 'wd_11pm_N', 'wd_11pm_NE']
for j in range(len(dummies_5am)):
  featurename = dummies_11pm[j]
  all_features[featurename] = all_features[dummies_5am[j]] - all_features[featurename]

```

```python
#Rename columns to adjust for above change
new_cols = []
for col in all_features.columns:
  if '11pm' in col:
    new_cols.append(col.replace('11pm', '6hrchange'))
  else:
    new_cols.append(col)

all_features.columns = new_cols
```

```python
#Add features containing spatial differences between adjacent model regions
#Count diagonal borders as adjacent
all_features.columns.values
first = ['3', '3', '3', '2', '2', '2', '2', '7', '7']
second = ['2', '5', '4', '4', '5', '6', '7', '4', '6']
fields = ['tsfc', 'tdsfc', 'usfc', 'vsfc', 'wssfc', 'psfc', 't800', 'td800', 'u800', 'v800', 'ws800']
for field in fields:
  for numidx in range(len(first)):
    all_features.insert(0, field + first[numidx] + second[numidx], all_features[field + first[numidx]] - all_features[field + second[numidx]])
```
